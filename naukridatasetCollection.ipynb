## Multiple url code

```python
import time
urls=[]
for i in range(100):
    elem=driver.find_elements(By.CLASS_NAME,'sjw__tuple')
    for ele in elem:
       url=ele.find_element(By.TAG_NAME,'a').get_property('href')
       if url not in urls:
           urls.append(url)
    print("collected url")
    driver.find_element(By.XPATH,'/html/body/div/div/main/div[1]/div[2]/div[3]/div/a[2]/span').click()
    time.sleep(10)
    print("next page loaded...")
    print()
```

```python
len(urls)
```

```python
urls
```

```python
with open("naukri.txt",mode='w') as file:
    for url in urls:
      file.write(f"{url}\n")
```

```python
with open("naukri.txt",mode='r') as file:
    prourls=[line for line in file.readlines() if line!='']
```

```python
urls=[url.strip("\n") for url in prourls]
```

```python
title=[]
company=[]
exp=[]
salary=[]
location=[]
posted=[]
applications=[]
role=[]
employment_type=[]
education=[]
skills=[]
for index in range(len(urls)):
    try:
        driver.get(urls[index])
        time.sleep(5)
        print(index,":",urls[index])
        print("succes")
        try:
            elements = driver.find_elements(By.CLASS_NAME, 'styles_jhc__jd-top-head__MFoZl')
            full_text = elements[0].text.strip()  # Get text from the first element
            title.append(full_text.split("\n")[0])
            company.append(full_text.split("\n")[1])
        except:
            title.append(np.nan)
            company.append(np.nan)   
        try:
            exp.append(driver.find_element(By.XPATH,'//*[@id="job_header"]/div[1]/div[2]/div[1]/div[1]/span').text)
        except:
            exp.append(np.nan)
        try:
            salary.append(driver.find_element(By.XPATH,'//*[@id="job_header"]/div[1]/div[2]/div[1]/div[2]/span').text) 
        except:
            salary.append(np.nan)
        try:
            location.append(driver.find_element(By.CLASS_NAME,'styles_jhc__loc___Du2H').text)
        except:
            location.append(np.nan)
        try:
           el12 = driver.find_elements(By.CLASS_NAME, 'styles_jhc__stat__PgY67')
           k=len(el12)
           if k==2:
                  full_text1 = el12[0].text
                  job_posted1= full_text1.split(":")[1]
                  full_text2 = el12[1].text
                  job_app2= full_text2.split(":")[1]
                  posted.append(job_posted1)
                  applications.append(job_app2)
           else:
                   full_text1 = el12[0].text
                   job_posted1= full_text1.split(":")[1]
                   full_text2 = el12[2].text
                   job_app2= full_text2.split(":")[1]
                   posted.append(job_posted1)
                   applications.append(job_app2)
        except:
            posted.append(np.nan)
            applications.append(np.nan)
        try:
           # Check if the "Read More" button exists before clicking
              read_more_buttons = driver.find_elements(By.CLASS_NAME, 'styles_read-more-link__dD_5h')
    
              if read_more_buttons:  # If button is found, click it
                  read_more_buttons[0].click()
                  time.sleep(2)  # Wait for content to expand
                  el = driver.find_elements(By.CLASS_NAME, 'styles_other-details__oEN4O')
                  full_text = el[0].text.strip()  # Get text from the first element
                  job_role = full_text.split("\n")[0].split(":")[1]
                  job_empl_type = full_text.split("\n")[3].split(":")[1]
                  employment_type.append(job_role)
                  role.append(job_empl_type)
              else:
                  el = driver.find_elements(By.CLASS_NAME, 'styles_other-details__oEN4O')
                  full_text = el[0].text.strip()  # Get text from the first element
                  job_role = full_text.split("\n")[0].split(":")[1]
                  job_empl_type = full_text.split("\n")[3].split(":")[1]
                  employment_type.append(job_role)
                  role.append(job_empl_type)

        except Exception as e:
             employment_type.append(np.nan)
             role.append(np.nan)
        try:
            el1 = driver.find_element(By.CLASS_NAME, 'styles_education__KXFkO')
            # Ensure at least one element exists
            full_text = el1.text.strip()  # Get text from the first element
            edu = full_text.replace('Education'," ")
            education.append(edu)
        except:
            education.append(np.nan)
        
        try:
            skills_elements = driver.find_element(By.CLASS_NAME, "styles_key-skill__GIPn_").find_elements(By.TAG_NAME, 'a')
            extracted_skills = [skill.text.strip() for skill in skills_elements if skill.text.strip()]
    
            # Convert list to a single comma-separated string and append to skills list
            formatted_skills = ", ".join(extracted_skills)
            skills.append(formatted_skills)  

        except:
              skills.append(np.nan)  
    except:
        print(index,":",urls[index])
        print("Not loaded")
        title.append(np.nan)
        company.append(np.nan)
        exp.append(np.nan)
        salary.append(np.nan)
        location.append(np.nan)
        posted.append(np.nan)
        applications.append(np.nan)
        role.append(np.nan)
        employment_type.append(np.nan)
        education.append(np.nan)
        skills.append(np.nan)
    print("Checking Lengths of Each List for Data Frame Conversion...")
    print(f" Title:{len(title)}, Company:{len(company)},Experience:{len(exp)},Salary:{len(salary)},Location:{len(location)},Posted:{len(posted)},Applications:{len(applications)},Role:{len(role)},Employment_type:{len(employment_type)},Education:{len(education)},Skills:{len(skills)}")
```

```python
df=pd.DataFrame({'Title':title,'Company':company,'Experience':exp,'Salary':salary,'Location':location,'Posted':posted,'Applications':applications,'Role':role,'Employment_type':employment_type,'Education':education,'Skills':skills})
```
