```python
def colvalidate(df, col):
    if df[col].dtype!=object:
        print(f"Column: {col}")
        print()
        print(f"Number of Unique Values: {df[col].nunique()}")
        print()
        print("Unique Values In Column:")
        # if values more than 100 , showing list wise unique values
        if df[col].nunique()>=100:
            for indx in range(0, df[col].nunique(), 100):
                print(df[col].unique()[indx:indx+100])
                print()
        else:
            print(df[col].unique())
            print()
        print("Data Type of Column:", df[col].dtype)
        print()
        print(f"Column: Min -> {df[col].min()} & Max -> {df[col].max()}")
        print("===============================================")
        print()
    else:
        print(f"Column: {col}")
        print()
        print(f"Number of Unique Values: {df[col].nunique()}")
        print()
        print("Unique Values In Column:")
        if df[col].nunique()>=100:
            for indx in range(0, df[col].nunique(), 100):
                print(df[col].unique()[indx:indx+100])
                print()
        else:
            print(df[col].unique())
            print()
        print("Data Type of Column:", df[col].dtype)
        print("===============================================")
        print()  
```

```python
import re
# List of all columns excluding the index
degree_categories = ["B.Tech/B.E", "B.Sc", "BCA", "M.Tech", "MBA", "MCA", "M.Sc", "Ph.D"]
ug_columns = ["B.Tech/B.E", "B.Sc", "BCA"]
pg_columns = ["M.Tech", "MBA", "MCA", "M.Sc", "Ph.D"]

# Step 2: Create new columns for each degree category and initialize with 0
for degree in degree_categories:
    df[degree] = 0

# Step 3: Process each row in the Education column
education_column = "Education"  # Define the correct column name
for i, row in df[education_column].dropna().items():  # Ignore NaN values
    for degree in degree_categories:
        if re.search(rf"\b{degree}\b", row, re.IGNORECASE):  # Check if degree is present
            df.at[i, degree] = 1 
for i, row in df[education_column].dropna().items():
    if "Any Graduate" and "Any Postgraduate"  in row:
        df.loc[i, degree_categories] = 1
    elif  "Any Graduate"  in row:
        df.loc[i,ug_columns] = 1
    elif "Any Postgraduate"  in row:
        df.loc[i,pg_columns] = 1
        # \nUG: B.Sc in Computers\nPG: MS/M.Sc(Science) in Data Informatics, Computers'
        #nUG: Any Graduate, Diploma in Mechanical\nPG: Any Postgraduate'

    
    

```

```python
df['Education']=df['Education'].str.strip(" \n").str.split("\n")
```

```python
for ind,col in df['Education'].items():
    col=str(col)
    if 'UG'  in col and 'PG' in col:
        df['Education'][ind]='Both'
    elif 'UG' in col:
        df['Education'][ind]='UG'
    elif 'PG' in col:
       df['Education'][ind]='PG'
```

```python
# Assuming skills are in a column named 'Skills'
skill_column = 'Skills'
df[skill_column] = df[skill_column].fillna('')

# Split skills into lists
df[skill_column] = df[skill_column].str.split(',')

# Standardized skill groups
skill_groups = {
    "Database": ["sql", "mysql", "postgresql", "sql server", "bigquery", "mongodb", "nosql", "dynamodb", "cloud sql"],
    "Testing": ["unit testing", "automation testing", "quality assurance", "selenium", "junit", "mockito", "soap ui"],
    "Programming": ["java", "python", "javascript", "rust", "typescript", "c#", "golang", "plsql", "scripting languages"],
    "Web Development": ["html", "css", "vue.js", "react.js", "node.js", "graphql", "angular", "rest apis", "django", "spring boot"],
    "Cloud Computing": ["gcp", "aws", "azure", "cloud infrastructure", "cloud storage", "cloud sql", "cloud technologies", "devops"],
    "Data-Driven Technologies": ["data science", "big data", "data analytics", "data engineering", "data mining", "machine learning", "deep learning", "ai/ml", "apache flink", "snowflake", "kafka"],
    "Project Management & DevOps": ["agile", "ci/cd methodology", "microservices", "terraform", "jenkins", "git"],
    "Analysis & BI": ["business analysis", "business intelligence", "qlikview"],
    "Mobile Development": ["react native", "flutter"],
    "Cybersecurity & API Management": ["api management", "security protocols", "oauth", "jwt"],
    "Containers & Orchestration": ["kubernetes", "docker", "hypervisor", "storage virtualization"],
}

# Function to map skills to their group
def categorize_skill(skill):
    skill = skill.strip().lower()
    for category, related_skills in skill_groups.items():
        if skill in related_skills:
            return category
    return skill  # Return skill as-is if no category match

# Create new columns for each skill group
for group in skill_groups.keys():
    df[group] = 0  # Initialize columns with 0

# Iterate over each row and update the skill group columns
for index, skills in enumerate(df[skill_column]):
    for skill in skills:
        skill_category = categorize_skill(skill)
        if skill_category in skill_groups:
            df.at[index, skill_category] = 1  # Mark presence of skill

# Drop original skills column if no longer needed
df = df.drop(columns=[skill_column])
```

```python

# Group by "Company" and "Title", sum the "Applications", and get the top 10
df_top = (
    df.groupby(["Company", "Title"])["Applications"].sum().nlargest(10).round(2).reset_index() ) # Convert Series to DataFrame)
# Display the top 10 values
display(df_top)

# Create the bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x="Company", y="Applications", hue="Title", data=df_top)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)
plt.title("Applications for Top 10 (Company, Title) Pairs")
plt.show()
```

```python

# Select relevant columns
degree_cols = ["B.Tech/B.E", "B.Sc", "BCA", "M.Tech", "MCA", "M.Sc", "MBA", "Ph.D", "Applications"]
melted_df = df[degree_cols].melt(id_vars=["Applications"], var_name="Degree", value_name="Required")
required_degrees = melted_df[melted_df["Required"] == 1]
print(required_degrees)
degree_counts = required_degrees.groupby("Degree")["Applications"].count().reset_index()
print(degree_counts)
degree_counts = degree_counts.sort_values("Applications", ascending=False)
plt.figure(figsize=(10, 6))
sns.barplot(x="Applications", y="Degree", data=degree_counts, palette="viridis")
plt.xlabel("Number of Applications")
plt.ylabel("Degree")
plt.title("Applications vs Degree Requirement")
plt.show()

```

```python

skill_cols = ['Database', 'Testing', 'Programming', 'Web Development', 'Cloud Computing',
              'Data-Driven Technologies', 'Project Management & DevOps', 'Analysis & BI',
              'Mobile Development', 'Cybersecurity & API Management', 'Containers & Orchestration', 'Applications']
melted_df = df[skill_cols].melt(id_vars=["Applications"], var_name="Skill", value_name="Required")
required_skills = melted_df[melted_df["Required"] == 1]
skill_counts = required_skills.groupby("Skill")["Applications"].count().reset_index()
skill_counts = skill_counts.sort_values("Applications", ascending=False)
plt.figure(figsize=(12, 6))
sns.barplot(x="Applications", y="Skill", data=skill_counts, palette="mako")
plt.xlabel("Number of Applications")
plt.ylabel("Skill")
plt.title("Applications vs Skills")
plt.show()
```
